```python

import os
import csv 
import numpy as np
import re 

##############################################################################################
## Define function for extracting CPU usage at different attack rates from generated files. ##
##############################################################################################
def get_raw_data(dir_path): 
    data_raw_unsorted = []

    for top_path in os.scandir(dir_path):
        content = [] 
        if top_path.is_dir():
            # dir name with extension
            full_dir_name = (os.path.basename(top_path))
            # print(full_dir_name)   
            rps=full_dir_name[4:]
            vect=[]
            for path in os.scandir(top_path): # rps_XXX directory 
                if path.is_file():
                    # file name with extension
                    full_file_name = (os.path.basename(path)) # outputX.txt file 
                    # print(full_file_name)    
                    file = open(dir_path+"/"+full_dir_name+"/"+full_file_name, mode='r') 
                    content = file.read().split('\n')[0:-1]  
                    for line in content: 
                        cpu_cycles=float(line.replace(',',''))
                        vect.append(cpu_cycles) 
                    file.close()
            
            #print("avg: ", avg)
            #print("count: ", count)
            data_raw_unsorted.append((float(rps), vect))
    data_raw_sorted = sorted(data_raw_unsorted)
    #print('latency data count:', len(data_raw_sorted))
    #print(data_raw_sorted[0])
    return data_raw_sorted


################################################################################
## Define functions for extracting flood rates and CPU cycles from raw data. ##
###############################################################################

def get_flood_rates(raw_data):
    flood_rate = []

    for file in raw_data:
        rate=file[0]
        flood_rate.append(rate)
    return flood_rate
    
def get_cpu_cycles(raw_data):
    cpu_cycles = []
    
    for file in raw_data:
        cycles=file[1]
        cpu_cycles.append(cycles)
    return cpu_cycles



#############################################################################################
## Define functions for scaling flood rates tp Mpps and CPU cycles to billions cycles/sec. ##
#############################################################################################

def get_scaled_flood_rates(flood_rate_unscaled): 
    flood_rate_scaled = [] 
    for r in flood_rate_unscaled: 
        flood_rate_scaled.append(float(r)/1e6) # flood rate in Mpps 
    return flood_rate_scaled

def get_scaled_avg_cpu_cycles(avg_cpu_cycles_unscaled):
    avg_cpu_cycles_scaled = []
    for c in avg_cpu_cycles_unscaled: 
        avg_cpu_cycles_scaled.append(float(c)/1e9)  # cycles in billions 
    return avg_cpu_cycles_scaled

    
```

```python
def parse_path(dir_path):
    raw_data = get_raw_data(dir_path)
    flood_rate = get_flood_rates(raw_data)
    cpu_cycles = get_cpu_cycles(raw_data)
    return flood_rate, cpu_cycles

def parse_and_plot(dir_path, color='r',style='-x'):
    x,yv= parse_path(dir_path)
    fx=[]
    fy=[]
    for i in range(len(x)):
        if (x[i]<15000)or (x[i]>100000):
            if x[i]<3.7e7:
                fx.append(x[i])
                fy.append(yv[i])
    fm=[np.mean(v) for v in fy]
    fs=[np.std(v) for v in fy]
    plt.plot(fx,fm,style,color=color)
    #plt.errorbar(fx,fm,yerr=fs,color=color)

plt.figure(figsize=(20,20))
parse_and_plot("data/cpu_instr/lucid/vanilla", color='black')

parse_and_plot("data/cpu_instr/lucid/jaqen_synflood", color='red')

parse_and_plot("data/cpu_instr/lucid/jaqen2_synflood", color='green')

parse_and_plot("data/cpu_instr/lucid/sc_synflood", color='grey')

parse_and_plot("data/cpu_instr/lucid/sc_ackflood", color='blue')

plt.grid(True,which='both')


plt.xlim([-0.2e7, 3.9e7])
plt.ylim([-0.1e8, 8.9e8])
```
