```python
# Bar plot. 

import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.1
fig = plt.subplots(figsize =(18, 8),dpi=300)
 
x_base = b_send_rate #already in Mpps 
y_base = b_return_ratio 

x_xdp = xdp_send_rate #already in Mpps 
y_xdp = xdp_return_ratio 

x_sc_rnd_port = sc_send_rate_rnd_port #already in Mpps 
y_sc_rnd_port = sc_return_ratio_rnd_port

x_jaqen = jaqen_send_rate #already in Mpps 
y_jaqen = jaqen_return_ratio 

x_aes = sc_aes_send_rate #already in Mpps 
y_aes = sc_aes_return_ratio 

x_crc = sc_crc_send_rate #already in Mpps 
y_crc = sc_crc_return_ratio 

ATTACK_RATES=[1, 5,10,15,20,30,40,50,75,100,150,200,220]
X_AXIS = np.arange(len(ATTACK_RATES))
 
plt.bar(X_AXIS-0.36, y_base, color ='k', width = barWidth, edgecolor='black', label ='Vanilla-SipHash')
plt.bar(X_AXIS-0.21, y_xdp, color ='b', width = barWidth, edgecolor='black', label ='XDP-SipHash')
plt.bar(X_AXIS-0.06, y_aes, color ='#7F00FF', width = barWidth, edgecolor='black',  label ='SC-AES')
plt.bar(X_AXIS+0.09, y_sc_rnd_port, color ='#00D100', width = barWidth, edgecolor='black', label ='SC-SipHash')
plt.bar(X_AXIS+0.24, y_jaqen, color ='r', width = barWidth, edgecolor='black', label ='Jaqen-CRC')
plt.bar(X_AXIS+0.39, y_crc, color ='c', width = barWidth, edgecolor='black',  label ='SC-CRC')

# Adding Xticks
plt.xlabel('SYN Flood Attack Rate (Mpps)', fontweight ='bold', fontsize = 25)
plt.ylabel('SYN-ACK Return Ratio (%)', fontweight ='bold', fontsize = 25)
plt.xticks(X_AXIS, ATTACK_RATES, fontsize=18)
plt.yticks(fontsize=18)

plt.ylim(0,1.2)
plt.legend(fontsize=19, loc='best', mode='expand', ncol=6)
plt.savefig('throughput_total.pdf')  

plt.show()
```
